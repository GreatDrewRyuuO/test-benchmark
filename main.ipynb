```python
import time
import statistics
import numpy as np
import matplotlib.pyplot as plt
```

```python
def slow_function():
    total = 0
    for i in range(10000):
        total += i
    return total


def faster_function():
    return sum(range(10000))
```

```python
def benchmark(func, iterations=1000):
    times = []
    for _ in range(iterations):
        start = time.perf_counter()
        func()
        end = time.perf_counter()
        times.append(end - start)
    return times
```

```python
slow_times = benchmark(slow_function)
fast_times = benchmark(faster_function)

print("Slow mean:", statistics.mean(slow_times))
print("Fast mean:", statistics.mean(fast_times))
```

```python
plt.figure()
plt.hist(slow_times, bins=30, alpha=0.6)
plt.hist(fast_times, bins=30, alpha=0.6)
plt.title("Performance Distribution")
plt.xlabel("Execution Time (seconds)")
plt.ylabel("Frequency")
plt.show()
```

```python
import timeit

slow_time = timeit.timeit(slow_function, number=1000)
fast_time = timeit.timeit(faster_function, number=1000)

print("Slow total:", slow_time)
print("Fast total:", fast_time)
```
